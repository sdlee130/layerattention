```python
from matplotlib import pyplot as plt

import numpy as np
import pandas as pd
```

```python
pivot = ['input', 'last', 'mean']
attn_func = ['dot', 'sdot', 'cos', 'gen']
size = ['large', 'base', 'mini']
```

```python
large_result = {'input': [76.0, 78.4, 78.6, 78.0],
               'last': [77.0, 78.4, 78.5, 77.1],
               'mean': [77.4, 77.9, 78.4, 78.3],
               'base': 76.4}

dot_values = [large_result[cat][0] for cat in pivot]
sdot_values = [large_result[cat][1] for cat in pivot]
cos_values = [large_result[cat][2] for cat in pivot]
gen_values = [large_result[cat][3] for cat in pivot]
```

```python
x = np.arange(len(pivot))
width = 0.1

min_value = 75
max_value = 80
```

```python
# Create the bar graph
fig, ax = plt.subplots(figsize=(10, 6))
bars_dot = ax.bar(x - 1.5 * width, dot_values, width, label='dot')
bars_sdot = ax.bar(x - 0.5 * width, sdot_values, width, label='sdot')
bars_cos = ax.bar(x + 0.5 * width, cos_values, width, label='cos')
bars_gen = ax.bar(x + 1.5 * width, gen_values, width, label='gen')

ax.axhline(y=large_result['base'], color='gray', linestyle='--', label='baseline')

# Adding labels and title
ax.set_xlabel('pivot layer')
ax.set_ylabel('glue score')
ax.set_xticks(x)
ax.set_xticklabels(pivot)
ax.legend()
ax.set_title('bert-large')
ax.set_ylim(min_value, max_value)

# Display the graph
plt.tight_layout()
plt.show()
```
